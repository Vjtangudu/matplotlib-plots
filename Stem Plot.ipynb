```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [2,4,3,5]
plt.stem(x,y,linefmt='--',markerfmt='ro',bottom=0,basefmt='green',label='python',use_line_collection=False,
         orientation='horizontal')
plt.legend()
plt.show()
```
