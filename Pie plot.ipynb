```python
import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ['python','c','c++','java']
plt.pie(x,labels=y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ['python','c','c++','java']
ex = [0.3,0.0,0.2,0.0]
plt.pie(x,labels=y,explode=ex)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ['python','c','c++','java']
plt.pie(x,labels=y,autopct='%0.1f%%',radius=0.7,startangle=180,labeldistance=1.1,rotatelabels=True)
        #shadow=True,textprops={'fontsize':10},wedgeprops={'linewidth':7,'edgecolor':'m'}counterclock=True
plt.title('Pie Chart')
plt.legend(loc=3)#loc=1 basic
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ['python','c','c++','java']
z = [20,30,10,40]
y1 = ['a','b','c','d']
plt.pie(x,labels=y,radius=1.5)
plt.pie(z,labels=y1,radius=1,startangle=90)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ['python','c','c++','java']
z = [1]
c = ['w']
plt.pie(x,labels=y,radius=1.5)
plt.pie(z,radius=1,colors=c)
plt.show()
```
