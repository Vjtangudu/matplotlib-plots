```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70]
plt.boxplot(x)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,120,130,230]
plt.boxplot(x,widths=0.1,labels=['Python'],showmeans=True,sym='b+'
           ,boxprops=dict(color='r'),capprops=dict(color='r'),whiskerprops=dict(color='r'),
           )#patch_artist=True,notch=True,vert=False,whis=2
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70]
x1 = [10,20,40,30,70,50,90]
y = [x,x1]
plt.boxplot(y,widths=0.1,labels=['Python','Java'],showmeans=True,sym='b+'
           ,boxprops=dict(color='r'),capprops=dict(color='r'),whiskerprops=dict(color='r'),
           )#patch_artist=True,notch=True,vert=False,whis=2

plt.show()
```
