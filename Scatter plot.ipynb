# Scatter plot

```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [2,4,6,8]
plt.scatter(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
days = [1,2,3,4]
number = [2,4,6,8]
c = ['r','g','b','y']
size = [100,150,200,250]
a = [0.8,0.6,0.4,0.2]
plt.scatter(days,number,color=c,s=size,alpha=a,marker="*",edgecolor='b',linewidth=0.8)
plt.title('Scatter Data')
plt.xlabel('Days')
plt.ylabel('Numbers')
plt.show()
```

```python
import matplotlib.pyplot as plt
days = [1,2,3,4]
number = [2,4,6,8]
number2 = [5,3,9,7]
c = ['r','g','b','y']
cc = ['g','y','r','b']
size = [100,150,200,250]
a = [0.8,0.6,0.4,0.2]
plt.scatter(days,number,color=c,s=size)
plt.scatter(days,number2,color=cc,s=size)
plt.title('Scatter Data')
plt.xlabel('Days')
plt.ylabel('Numbers')
plt.show()
```
