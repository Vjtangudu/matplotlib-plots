```python
#fill between
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])

plt.plot(x,y)
plt.fill_between(x,y,color='green',where=(x>=2)&(y<=4),alpha=0.5)
plt.show()
```

# sub-plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[1,2,3,4]
plt.subplot(2,2,1)
plt.plot(x,y,color='r')

z=[50,30,20]
z1=[50,30,20]
plt.subplot(2,2,2)
plt.pie(z,labels=z1,explode=[0.0,0.2,0.0],colors=['r','b','g'])

x1=[1,2,3,4]
y1=[1,2,3,4]
plt.subplot(2,2,3)
plt.pie(x1,labels=y1)

x2=['v','i','j','a','y']
y2=[10,20,30,40,50]
plt.subplot(2,2,4)
plt.bar(x2,y2)

plt.show()
```
