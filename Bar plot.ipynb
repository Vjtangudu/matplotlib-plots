# matplotlib

```python
from matplotlib import pyplot as plt
x = [1,2,3,4]
y = [2,4,6,8]
plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4]
y = [2,4,6,8]
c = ['r','b','g','y']
plt.bar(x,y,color=c)
plt.show()
```

```python
x = ['python','c','c++','java']
y = [2,4,6,8]
plt.xlabel('Languages',fontsize=15)
plt.ylabel('Numbers',fontsize=15)
plt.title('LxN',fontsize=15)
c=['r','y','b','g']
plt.bar(x,y,width=0.4,color=c,align='edge',edgecolor='r',linewidth=5,linestyle=':',alpha=0.5)#align can have edge and center
plt.show()
```

```python
x = ['python','c','c++','java']
y = [2,4,6,8]
plt.xlabel('Languages',fontsize=15)
plt.ylabel('Numbers',fontsize=15)
plt.title('LxN',fontsize=15)
c=['r','y','b','g']
plt.bar(x,y,width=0.4,color='r',align='edge',edgecolor='r',label='vijay')
plt.legend()
plt.show()
```

```python
x = ['python','c','c++','java']
y = [2,4,6,8]
z = [1,2,3,4]
plt.xlabel('Languages',fontsize=15)
plt.ylabel('Numbers',fontsize=15)
plt.title('LxN',fontsize=15)
plt.bar(x,y,width=0.4,color='r',label='popularity')
plt.bar(x,z,width=0.4,color='y',label='popularity1')
plt.legend()
plt.show()
```

```python
import numpy as np
x = ['python','c','c++','java']
y = [2,4,6,8]
z = [1,2,3,4]
width = 0.2
p = np.arange(len(x))
p1 = [j+width for j in p]
plt.xlabel('Languages',fontsize=15)
plt.ylabel('Numbers',fontsize=15)
plt.title('LxN',fontsize=15)
plt.bar(p,y,width,color='r',label='popularity')
plt.bar(p1,z,width,color='y',label='popularity1')
plt.xticks(p+width/2,x,rotation=10)
plt.legend()
plt.show()
```

```python
import numpy as np
x = ['python','c','c++','java']
y = [2,4,6,8]
z = [1,2,3,4]
width = 0.2
p = np.arange(len(x))
p1 = [j+width for j in p]
plt.xlabel('Languages',fontsize=15)
plt.ylabel('Numbers',fontsize=15)
plt.title('LxN',fontsize=15)
plt.barh(p,y,width,color='r',label='popularity')
plt.barh(p1,z,width,color='y',label='popularity1')
plt.yticks(p+width/2,x,rotation=10)
plt.legend()
plt.show()
```

```python
x = ['python','c','c++','java']
y = [2,4,6,8]
z = [1,2,3,4]
plt.ylabel('Languages',fontsize=15)
plt.xlabel('Numbers',fontsize=15)
plt.title('LxN',fontsize=15)
plt.barh(x,y,color='r',label='popularity')
plt.barh(x,z,color='y',label='popularity1')
plt.legend()
plt.show()
```
