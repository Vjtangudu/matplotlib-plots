```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
area1=[3,5,3,6,5]
area2=[2,4,2,5,4]
area3=[1,2,2,4,1]
l=['area1','area2','area3']
plt.stackplot(x,area1,area2,area3,labels=l,colors=['g','y','r'])
plt.legend(loc=2)
plt.show()
```
